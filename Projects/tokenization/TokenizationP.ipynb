```python
import re
f=open("mycode.txt", "r") #Reading  a file
if f.mode == 'r': #checking if the file is in read state
    contents =f.read()
    mycode  = contents.split()
    keywords = re.compile(r'Until|Var|Check|Un_check|Show|Get|Func|Main|Load')
    identifier = re.compile(r'^[^\d\W]\w*\Z')
    o_bracket = re.compile(r'[(|{]')
    c_bracket = re.compile(r'[)|}]')
    terminator = re.compile(r';')
    operators = re.compile(r'[+ | ++ | -- | == | * | / | % ]')
    literals = re.compile(r'''(\"\"\"|\'\'\'|\"|\')((?<!\\)\\\1|.)*?\1''')
    digits = re.compile(r'[\d{9,10}]')
    
    for i in mycode:
        iden_re = re.search(identifier, i) 
        close_re = re.search(c_bracket,i)
        open_re = re.search(o_bracket,i)
        terminator_re= re.search(terminator,i)
        operators_re = re.search(operators,i)
        literals_re = re.search(literals,i)
        digits_re =re.search(digits,i)
        keywords_re = re.search(keywords,i)
        if keywords_re!=None:
            print("Keyword: ",i)
        elif iden_re!=None:
            print("Identifier: ",i )
        elif close_re!=None:
            print("close: ",i  )
        elif open_re!=None:
            print("Open Brackets: ",i )
        elif terminator_re!=None:
            print("Terminator: ",i )
        elif operators_re!=None:
            print("Operator: ",i )
        elif literals_re!=None:
            print("Literals: ",i )
        elif digits_re!=None:
            print("Digits: ",i )
```
