```python
import re as regex

import re
def token():
    f=open("code.txt", "r") #Reading  a file
    if f.mode == 'r': #checking if the file is in read state
        contents =f.read()
        mycode  = contents.split()
        keywords = re.compile(r'until|var|check|uncheck|show|get|main')
        identifier = re.compile(r'^[^\d\W]\w*\Z')
        o_bracket = re.compile(r'[(|{]')
        c_bracket = re.compile(r'[)|}]')
        terminator = re.compile(r';')
        operators = re.compile(r'[+ | ++ | -- | == | * | / | % ]')
        literals = re.compile(r'''(\"\"\"|\'\'\'|\"|\')((?<!\\)\\\1|.)*?\1''')
        digits = re.compile(r'[\d{9,10}]')

        for i in mycode:
            iden_re = re.search(identifier, i) 
            close_re = re.search(c_bracket,i)
            open_re = re.search(o_bracket,i)
            terminator_re= re.search(terminator,i)
            operators_re = re.search(operators,i)
            literals_re = re.search(literals,i)
            digits_re =re.search(digits,i)
            keywords_re = re.search(keywords,i)
            if keywords_re!=None:
                print("Keyword: ",i)
            elif iden_re!=None:
                print("Identifier: ",i )
            elif close_re!=None:
                print("close: ",i  )
            elif open_re!=None:
                print("Open Brackets: ",i )
            elif terminator_re!=None:
                print("Terminator: ",i )
            elif operators_re!=None:
                print("Operator: ",i )
            elif literals_re!=None:
                print("Literals: ",i )
            elif digits_re!=None:
                print("Digits: ",i )
token()


def get_code():
    file = open("code.txt", "r");
    # code=file.readlines()
    code = file.read()
    # code1=file.readline()
    # print(code)
    return code


code = get_code()
code_exe = ""
count = 0

for i in code:
    # print(i)
    if (i == "\n" or i == " "):
        continue
    else:
        code_exe = code_exe + i
        count += count

print(code_exe)

meaning_block = ""
meaning_temp = ""
count1 = 0
count2 = 0
count3 = 0

show = regex.compile(r'show\(\"\w+\"\);')
get = regex.compile(r'get\(\"\w+\"\);')
vari = regex.compile(r'([a-z]+[A-Z0-9]*=\d+)|([a-z]+[A-Z0-9]*=\"\w+\")')
check = regex.compile(r'check\((\w*|\d*|[=]|\d*)\)({(\s+|\S+|\w+|\d+)}|[{][}])')
check_=regex.compile(r'check_\((\w*|\d*|[=]|\d*)\)({(\s+|\S+|\w+|\d+)}|[{][}])')
uncheck=regex.compile(r'uncheck({(\s+|\S+|\w+|\d+)}|[{][}])')
untill=regex.compile(r'untill\(;\w+[=<>]\d+;\w+([+]+|[-]+)\)[{][}]')
main=regex.compile(r'main\(\){}')

load = regex.compile(r'load[a-z]+[A-Z0-9]*')

for i in code_exe:
    meaning_block = meaning_block + i
    print(meaning_block)
    print("\n")

    if (get.findall(meaning_block)):
        count1 = count1 + 1
        temp = get.search(meaning_block)
        # print("Span")
        # print(temp.span())
        temp = temp.span()
        # print(temp[0])
        # print(temp[1])

        meaning_temp = meaning_block[int(temp[0]):int(temp[1])]
        meaning_block = meaning_block.replace(meaning_temp, "")

        print("------------------------------------------------------------")
        print("\n Detected " + meaning_temp + "\n")
        print(" Get Method Matched")

        print("------------------------------------------------------------")
        # print(meaning_block)

        # print(count1)

    elif (vari.findall(meaning_block)):
        count1 = count1 + 1
        temp = vari.search(meaning_block)
        temp = temp.span()
        meaning_temp = meaning_block[int(temp[0]):int(temp[1])]
        meaning_block = meaning_block.replace(meaning_temp, "")

        print("------------------------------------------------------------")
        print("\n Detected " + meaning_temp + "\n")
        print(" Variables Matched")

        print("------------------------------------------------------------")
        # print(meaning_block)

        # print(count1)
    elif (check.findall(meaning_block)):
        count1 = count1 + 1
        temp = check.search(meaning_block)
        temp = temp.span()
        meaning_temp = meaning_block[int(temp[0]):int(temp[1])]
        meaning_block = meaning_block.replace(meaning_temp, "")

        print("------------------------------------------------------------")
        print("\n Detected " + meaning_temp + "\n")
        print(" Condition Matched")

        print("------------------------------------------------------------")
        # print(meaning_block)

        # print(count1)

    elif (check_.findall(meaning_block)):
        count1 = count1 + 1
        temp = check_.search(meaning_block)
        temp = temp.span()
        meaning_temp = meaning_block[int(temp[0]):int(temp[1])]
        meaning_block = meaning_block.replace(meaning_temp, "")

        print("------------------------------------------------------------")
        print("\n Detected " + meaning_temp + "\n")
        print(" Conditional else if Matched")

        print("------------------------------------------------------------")
        # print(meaning_block)

        # print(count1)
    elif (uncheck.findall(meaning_block)):
        count1 = count1 + 1
        temp = uncheck.search(meaning_block)
        temp = temp.span()
        meaning_temp = meaning_block[int(temp[0]):int(temp[1])]
        meaning_block = meaning_block.replace(meaning_temp, "")

        print("------------------------------------------------------------")
        print("\n Detected " + meaning_temp + "\n")
        print(" Condition else Matched")

        print("------------------------------------------------------------")
        # print(meaning_block)

        # print(count1)

    elif (untill.findall(meaning_block)):
        count1 = count1 + 1
        temp = untill.search(meaning_block)
        temp = temp.span()
        meaning_temp = meaning_block[int(temp[0]):int(temp[1])]
        meaning_block = meaning_block.replace(meaning_temp, "")

        print("------------------------------------------------------------")
        print("\n Detected " + meaning_temp + "\n")
        print(" Loop Matched")

        print("------------------------------------------------------------")
        # print(meaning_block)

    elif (main.findall(meaning_block)):
        count1 = count1 + 1
        temp = main.search(meaning_block)
        temp = temp.span()
        meaning_temp = meaning_block[int(temp[0]):int(temp[1])]
        meaning_block = meaning_block.replace(meaning_temp, "")
        print("------------------------------------------------------------")
        print("\n Detected " + meaning_temp + "\n")
        print(" Program Entry Point ")
        print("------------------------------------------------------------")
        # print(meaning_block)




    elif (show.findall(meaning_block)):
        count1 = count1 + 1
        temp = show.search(meaning_block)
        # print("Span")
        # print(temp.span())
        temp = temp.span()
        # print(temp[0])
        # print(temp[1])

        meaning_temp = meaning_block[int(temp[0]):int(temp[1])]
        meaning_block = meaning_block.replace(meaning_temp, "")

        print("------------------------------------------------------------")
        print("\n Detected " + meaning_temp + "\n")
        print(" Show Method Matched")

        print("------------------------------------------------------------")
        # print(meaning_block)

        # print(count1)

    else:
        continue
        # count1=count1+1



#print(len(meaning_block))
if(len(meaning_block)!=0):
    print("Error In Your Code")
else:
    print("Program Successful")
```
